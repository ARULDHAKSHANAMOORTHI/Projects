```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOADING DATASET

```python
df=pd.read_csv("/netflix_titles.csv")
df.head()
```

# EXPLORATORY DATA ANALYSIS

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.dropna()
df.shape
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
df['day_added']=df['date_added'].dt.day
df['month_added']=df['date_added'].dt.month
df['year_added']=df['date_added'].dt.year
```

```python
df.dtypes
```

```python
movie=df[df['type']=='Movie']
tv=df[df['type']=='TV Shows']
```

# DATA VISUALISATION

```python
df.type.value_counts().plot(kind='pie')
plt.title("MOSTLY WATCHED SHOW ON NETFLIX")
plt.legend()
plt.show()
```

```python
df.director.value_counts().head()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=df)
plt.title("Plot of rating by type")
plt.show()
```

```python
df.country.value_counts()
```

```python
plt.figure(figsize=(10,8))
df.country.value_counts().head(10).plot(kind='bar',color='blue')
plt.title("Top 10 countries releases show on Netflix")
plt.xlabel("COuntries")
plt.ylabel("COunts")
plt.show()
```

```python
plt.figure(figsize=(10,8))
movie.release_year.value_counts().head(10).plot(kind='bar',color='magenta')
plt.title('Top 10 years in which most movies are released')
plt.xlabel('Release Year')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
df['month_added'].value_counts().head(10).plot(kind='bar')
plt.title("Monthwise show release")
plt.legend()
plt.xlabel("Months")
plt.show()
```

```python
df['day_added'].value_counts().head(10).plot(kind='bar')
plt.title("Daywise show release")
plt.legend()
plt.xlabel("Days")
plt.show()
```
